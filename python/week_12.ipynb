```python
import csv
f = open('seoul.csv', 'r',encoding = 'cp949')
data = csv.reader(f, delimiter=",")
header = next(data)
change_temp = -999
max_temp = -999
min_temp = 999
change_date = ''
max_date = ''
min_date = ''
for row in data:
    if row[4] == '':
        row[4] = -999
    row[3] = float(row[3])
    row[4] = float(row[4])
    if change_temp < row[4] - row[3]:
        change_date = row[0]
        change_temp = row[4] - row[3]
    if min_temp > row[3]:
        min_date = row[0]
        min_temp = row[3]
    if max_temp < row[4]:
        max_date = row[0]
        max_temp = row[4]
print('일교차가 가장 컸던 날')
print('날짜',change_date)
print('최저기온이 가장 낮은 날')
print('날짜',min_date)
print('최고기온이 가장 높은 날')
print('날짜',max_date)
f.close()
```

```python
import csv
f = open('subway.csv', 'r',encoding = 'utf-8')
data = csv.reader(f, delimiter=",")
header = next(data)
dic = {}
for row in data:
    row[4] = row[4].replace(',','')
    row[5] = row[5].replace(',','')
    row[4] = int(row[4])
    row[5] = int(row[5])
    if dic.get(row[1]):
        dic[row[1]][row[3]] = row[4] + row[5]
    else:
        dic[row[1]] = {row[3]:row[4] + row[5]}
f.close()
for k, v in dic.items():
    v = sorted(v.items(), key = lambda item: item[1])
    print(k)
    print('이용객이 가장 많은 역 순위')
    print('1.',v[-1][0],'(',v[-1][1],')')
    print('2.',v[-2][0],'(',v[-2][1],')')
    print('3.',v[-3][0],'(',v[-3][1],')')
    print('이용객이 가장 적은 역 순위')
    print('1.',v[1][0],'(',v[1][1],')')
    print('2.',v[2][0],'(',v[2][1],')')
    print('3.',v[3][0],'(',v[3][1],')')
```
